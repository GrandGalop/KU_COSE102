```python
inputstr=input()
qwer=["q", "w", "e", "r", "t", "y", "u", "i", "o", "p"]
asdf=["a", "s", "d", "f", "g", "h", "j", "k", "l"]
zxcv=["z", "x", "c", "v", "b", "n", "m"]
pos=[]
for i in range(0, len(inputstr)):
  if inputstr[i] in qwer:
    pos.append(4)
  elif inputstr[i]in asdf:
    pos.append(3)
  elif inputstr[i]in zxcv:
    pos.append(2)
  else:
    pos.append(1)
tired=0
for i in range(0, len(pos)-1):
  tired = tired+abs(pos[i+1]-pos[i])
print(tired)
```
