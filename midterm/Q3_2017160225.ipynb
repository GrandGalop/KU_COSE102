```python
firstlist=input().split()
secondlist=input().split()
for i in range(0, len(firstlist)):
  firstlist[i]=int(firstlist[i])

for i in range(0, len(secondlist)):
  secondlist[i]=int(secondlist[i])

firstlist.sort()
secondlist.sort()
if firstlist==secondlist:
  print("YES")
else:
  print("NO")
```
